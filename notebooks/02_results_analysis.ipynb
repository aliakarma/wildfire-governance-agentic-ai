# Results Analysis — Reproduce Paper Figures

Loads pre-committed CSVs from `results/paper/` and regenerates all paper figures.
No experiments need to be run — figures are generated from committed data.

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from wildfire_governance.utils.visualisation import (
    apply_ieee_style, plot_latency_vs_uavs,
    plot_false_alert_bar, plot_tradeoff_frontier
)

apply_ieee_style()
PAPER = Path('../results/paper')
print('Paper CSVs available:', [f.name for f in PAPER.glob('*.csv')])
```

```python
# Table II — RL Policy Comparison
t2 = pd.read_csv(PAPER / 'table2_rl_comparison.csv')
print('Table II — RL Policy Comparison (N=20 UAVs, 20 seeds):')
print(t2[['method','framework','ld_mean','ld_std','fp_mean','fp_std','governance_compliance_pct']].to_string(index=False))
```

```python
# Figure 3 — Detection Latency vs. Fleet Size
fig3_data = pd.read_csv(PAPER / 'fig3_latency_data.csv')
fig3_data = fig3_data.rename(columns={'config': 'config'})
fig = plot_latency_vs_uavs(fig3_data, show=True)
plt.title('Fig. 3 — Detection Latency vs. UAV Fleet Size')
```

```python
# Figure 5 — Latency-False Alert Tradeoff (N=40)
fig5_data = pd.read_csv(PAPER / 'fig5_tradeoff_data.csv')
fig5_data = fig5_data.rename(columns={'config': 'config'})
fig = plot_tradeoff_frontier(fig5_data, show=True)
plt.title('Fig. 5 — Latency-False Alert Tradeoff (N=40 UAVs)')
```

```python
# Table V — Adversarial Robustness
t5 = pd.read_csv(PAPER / 'table5_adversarial.csv')
print('Table V — Adversarial Robustness:')
print(t5[['attack_type','parameter','gomdp_fp','central_fp','p_breach_gomdp']].to_string(index=False))
```
